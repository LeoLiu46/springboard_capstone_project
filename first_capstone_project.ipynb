### Replace specific values in categorical and numerical features to nan

```python
# replace all 999998 to nan, 999998 stands for 'logical skip' under NOCPR(Job code for principal job)
df_nan_999998 = df.replace(999998, np.nan)
df_nan_999998.isnull().sum().sum()
```

```python
# replace all 9999998 to nan, 9999998 stands for 'logical skip' under SALARY
df_nan_999998_9999998 = df_nan_999998.replace(9999998, np.nan)
df_nan_999998_9999998.isnull().sum().sum()
```

```python
# replace all 98 to nan, 98 stands for 'logical skip' under 106 features(for details see window below)  55-3+26+7+14
df_nan_999998_9999998_98 = df_nan_999998_9999998.replace(98, np.nan)
df_nan_999998_9999998_98.isnull().sum()
```

#any column here with a nan value is a categorical feature which needs to be dummied later

### Splitting the dataset into numerical part(part1) and categorical part(part2)

```python
#Selecting all numerical features and save them to df_merge_part1
df_merge_part1 = df_nan_999998_9999998_98[['PERSONID','AGE','BA03Y5','HD03Y5','SALARY']]
df_merge_part1.isnull().sum()
```

```python
#Converting 'Year of bachelor degree 2003-onward (5 year intervals)' to 'Years after bachelor degree 2003-onward (5 year intervals)'
df_merge_part1['BA03Y5'] = 2010 - df_merge_part1['BA03Y5']
```

```python
#Converting 'Year of highest degree 2003-onward (5 year intervals)' to 'Years after highest degree 2003-onward (5 year intervals)'
df_merge_part1['HD03Y5'] = 2010 - df_merge_part1['HD03Y5']
```

```python
df_merge_part1.head()
```

```python
#Selecting the categorical part
df_merge_part2_nan_999998_9999998_98 = df_nan_999998_9999998_98.drop\
(['AGE','BA03Y5','HD03Y5','SALARY','YEAR','WEIGHT','SAMPLE','SURID'], axis = 1)

df_merge_part2_nan_999998_9999998_98.shape
```

### Converting all categorical features to dummy variables 

```python
df_merge_part2_nan_999998_9999998_98.info()
```

```python
df_merge_part2_nan_999998_9999998_98.shape
```

```python
df_merge_part2 = pd.get_dummies(df_merge_part2_nan_999998_9999998_98,\
                                columns = [col for col in df_merge_part2_nan_999998_9999998_98.columns][1:],\
                                dummy_na = False)
df_merge_part2.shape
```

```python
df_merge_part2.head()
```

### Merging numerical features and categorical features

```python
df_merge = df_merge_part1.merge(df_merge_part2, on = 'PERSONID', how = 'left').dropna(how = 'any')
df_merge.shape
```

```python
#export clean dataset
df_merge.to_csv('D:/Work_Study/leo_ongoing_project/springboard_data_scientist_track/springboard_capstone_project/data_metadata/df_merge.csv',\
                index = False)
```
